```python
import pandas as pd
from datetime import datetime, timedelta
```

```python
sentiment_data = "../SPRING/labelled_data.csv"
emotion_data = "../SUMMER/labelled_data.csv"

with open(sentiment_data, "r") as s, open(emotion_data, "r") as e:
    senti_df = pd.read_csv(s)
    emo_df = pd.read_csv(e)
```

```python
emo_df
```

```python
# Check if tweet columns are equal
eq_cols = ["date","tweet_id","processed_tweets"]
print(senti_df[eq_cols].equals(emo_df[eq_cols]))
```

```python
# Merge columns
label_df = emo_df.merge(senti_df, on = eq_cols, how = "inner")
```

```python
# Make sure that all tweets are present
print(label_df[eq_cols].equals(senti_df[eq_cols]))
print(label_df[eq_cols].equals(emo_df[eq_cols]))
```

```python
translations = "../SUMMER/stats/data/names-and-populations.csv"
with open(translations, "r") as f:
    translations = pd.read_csv(f)[["country","name_en","name_translated","population"]]
translations
```

```python
label_df = pd.merge(left = label_df, right = translations, left_on=["city"], 
                  right_on = ["name_translated"],how = "left")
label_df = label_df.drop(["name_translated","population"], axis=1).rename(columns={"name_en":"city_en",
                                                                "city":"city_ar"})
label_df["city_en"] = label_df["city_en"].str.lower()
label_df["country"] = label_df["country"].str.lower()
label_df
```

```python
# Make sure that all tweets are present
print(label_df[eq_cols].equals(senti_df[eq_cols]))
print(label_df[eq_cols].equals(emo_df[eq_cols]))
```

```python
regs = "data/Sudan_city_regions_2023-10-09.csv"
with open(regs, "r") as f:
    regions = pd.read_csv(f)
regions["name_en"] = regions["name_en"].str.lower()
regions["Sudan_regions"] = regions["Sudan_regions"].str.lower()
regions
```

```python
regions_dict = regions.set_index("name_en")["Sudan_regions"].to_dict()

def convert_to_sudan_region(val):
    try:
        return regions_dict[val].lower()
    except KeyError:
        return val.lower()

label_df["city_en"] = label_df["city_en"].apply(convert_to_sudan_region)
label_df.loc[label_df["country"]=="south sudan", "city_en"] = "south sudan"
with open("data/senti_emo_labels.csv","w+") as o:
    cols = ["date","country","city_ar","city_en","tweet_id","raw_tweet","processed_tweets","language",
            "anger_NaiveBayes","fear_NaiveBayes","sadness_NaiveBayes","disgust_NaiveBayes","joy_NaiveBayes","anger-disgust_NaiveBayes",
            "anger_Lex","fear_Lex","sadness_Lex","disgust_Lex","joy_Lex","anger-disgust_Lex",
            "anger_GloVe","fear_GloVe","sadness_GloVe","disgust_GloVe","joy_GloVe","anger-disgust_GloVe",
            "sentiment_pos_comboLex","sentiment_neg_comboLex","sentiment_neu_comboLex",
            "sentiment_SVM","sentiment_pos_multiBERT","sentiment_neg_multiBERT"]
    label_df[cols].sort_values("date").to_csv(o, index=False)
label_df
```

```python
# Make sure that all tweets are present
print(label_df[eq_cols].equals(senti_df[eq_cols]))
print(label_df[eq_cols].equals(emo_df[eq_cols]))
```

```python
iom = "IOM/combined_4_27_8_10.csv"
with open(iom, "r") as f:
    iom_df = pd.read_csv(f).fillna(0)
    iom_df["location"] = iom_df["location"].str.lower()
    iom_df["date"] = pd.to_datetime(iom_df["date"], format='%m/%d/%Y').dt.strftime('%Y-%m-%d')
iom_df
```

```python
iom_dates = sorted(list(iom_df["date"].unique()))
iom_dates
```

```python
print(sorted(list(label_df["date"].unique())))
```

```python
date_range = [datetime.strptime(s, "%m/%d/%Y") for s in ["04/20/2023","08/10/2023"]]
iom_dates = [datetime.strptime(d, "%Y-%m-%d") for d in iom_dates]
tweet_dates = sorted([datetime.strptime(d, "%Y-%m-%d") for d in label_df["date"].unique()])

while date_range[-1] > tweet_dates[0]:
    date_range.append(date_range[-1] - timedelta(days = 7))
while date_range[-1] < tweet_dates[-1]:
    date_range.append(date_range[-1] + timedelta(days = 7))
date_range = sorted(date_range)

def date_later(val):
    date = datetime.strptime(val, "%Y-%m-%d")
    for d in date_range:
        if date <= d:
            return datetime.strftime(d, "%Y-%m-%d")


label_df["date"] = label_df["date"].apply(date_later)
print(sorted(list(label_df["date"].unique())))
```

```python
label_df = pd.merge(left = label_df, right = iom_df, right_on = ["date","location"],
                    left_on = ["date","city_en"], how = "left")
label_df = label_df.drop("location", axis=1)
```

```python
# Make sure that all tweets are present
print(label_df[eq_cols].equals(senti_df[eq_cols]))
print(label_df[eq_cols].equals(emo_df[eq_cols]))
```

```python
col_order = ["date","country","city_en","city_ar",'arriving_IDP', 'leaving_IDP',"tweet_id","raw_tweet",
            "processed_tweets","language",'anger_NaiveBayes','fear_NaiveBayes', 'sadness_NaiveBayes', 
            'disgust_NaiveBayes','joy_NaiveBayes', 'anger-disgust_NaiveBayes', 'anger_Lex', 'fear_Lex',
            'sadness_Lex', 'disgust_Lex', 'joy_Lex', 'anger-disgust_Lex','anger_GloVe', 'fear_GloVe', 'sadness_GloVe', 
            'disgust_GloVe','joy_GloVe', 'anger-disgust_GloVe','sentiment_pos_comboLex', 'sentiment_neg_comboLex',
            'sentiment_neu_comboLex', 'sentiment_SVM', 'sentiment_pos_multiBERT','sentiment_neg_multiBERT']
label_df = label_df[col_order]
label_df
```
